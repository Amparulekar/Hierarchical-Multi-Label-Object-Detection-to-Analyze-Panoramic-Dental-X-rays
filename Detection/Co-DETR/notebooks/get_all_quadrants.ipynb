```python
# CHANGE
# # for uncompressed
# folder = "/raid/infolab/bhavyakohli/mmdet/mmdetection/data/training_data/quadrant_enumeration/xrays/"

# for compressed 2
folder = "/raid/infolab/bhavyakohli/mmdet/mmdetection/data/training_data/compressed2/quadrant_enumeration/xrays/"

# # for compressed 3
# folder = "/raid/infolab/bhavyakohli/mmdet/mmdetection/data/training_data/Compressed_training_data/quadrant_enumeration/xrays/"

# # for compressed 4
# folder = "/raid/infolab/bhavyakohli/mmdet/mmdetection/data/training_data/compressed4/quadrant_enumeration/xrays/"


images = [folder + i['file_name'] for i in f['images']]

annotations = []

for n, image in enumerate(tqdm(images)):
    result = inference_detector(model, image)
    bboxes = [i[0][:-1] for i in result]
    img_id = f['images'][n]['id']

    for cat_id, bbox in enumerate(bboxes):
        annot = {}
        annot['image_id'] = img_id
        annot['category_id'] = cat_id
        annot['bbox'] = [float(i) for i in list(bbox)]
        annot['id'] = 4*n + cat_id + 1

        annotations.append(annot)
```

```python
for i in range(len(annotations)):
    annotations[i]['bbox'] = [float(j) for j in annotations[i]['bbox']]
```

```python
final_annotations = {'images': f['images'], 'annotations': annotations, 'categories': target['categories']} 

# CHANGE
# # for uncompressed
# json.dump(final_annotations, open("/raid/infolab/bhavyakohli/mmdet/Co-DETR/data/enumeration/quadrants_of_enumeration_data.json", 'w'))

# for compressed 2
json.dump(final_annotations, open("/raid/infolab/bhavyakohli/mmdet/Co-DETR/data/wave2_enum/quadrants_of_enumeration_data.json", 'w'))

# # for compressed 3
# json.dump(final_annotations, open("/raid/infolab/bhavyakohli/mmdet/Co-DETR/data/wave3_enum/quadrants_of_enumeration_data.json", 'w'))

# # for compressed 4
# json.dump(final_annotations, open("/raid/infolab/bhavyakohli/mmdet/Co-DETR/data/wave4_enum/quadrants_of_enumeration_data.json", 'w'))

```
